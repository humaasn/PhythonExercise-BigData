# Filtering
- 특정 조건식을 만족하는 Data만 따로 추출하는 방법

## Boolean indexing
- True 와 False 값으로 조건에 해당하는 값 찾기

```python
# 나이가 10대(10~19)인 승객만 선택
mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic.loc[mask1,:]
df_teenage.head()
```

```python
# 나이가 10세 미만(0 ~ 9세)이고 여성인 승객만 따로 선택
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2,:]
print(len(df_female_under10),'개')
df_female_under10.head()
```

```python
# 나이가 10세 미만 또는 60세 이상인 승객의 age, sex, alone 열만 선택하기
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3,['age','sex','alone']]
print(len(df_under10_morethan60),'개')
df_under10_morethan60.head()
```

### isin()
- 특정값을 가진 행들을 따로 추출

```python
# Boolean Indexing 으로 함께 탑승한 형제 또는 배우ㅏ자의 수가 3,4,5 인 승객만 추출
mask3 = titanic['sibsp'] == 3
mask4 = titanic['sibsp'] == 4
mask5 = titanic['sibsp'] == 5

df_boolean = titanic[mask3 | mask4 | mask5]
print(len(df_boolean),'개')
df_boolean.head()
```

```python
# isin() 매서드를 활용하여 동일한 조건을 추출
isin_filter = titanic['sibsp'].isin([3,4,5])

df_isin = titanic[isin_filter]
print(len(df_isin),'개')
df_isin.head()
```

# DataFrame 합치기
- 데이터가 여러 군데 나누어져 있을때 하나로 합치거나 데이터를 연결하는 경우

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3'],
                    'c':['c0','c1','c2','c3']},
                   index=[0,1,2,3]
                  )
df1
```

```python
df2 = pd.DataFrame({'a':['a2','a3','a4','a5'],
                    'b':['b2','b3','b4','b5'],
                    'c':['c2','c3','c4','c5'],
                    'd':['d2','d3','d4','d5']},
                   index=[2,3,4,5]
                  )
df2
```

```python
# 2개의 데이터 프레임을 위 아래로 행 방향으로 이어 붙이듯 연결하기 - r로 치면 rbind
result1 = pd.concat([df1,df2])
result1
```

```python
result2 = pd.concat([df1,df2],ignore_index=True) # ignore_index - index 재배치 
result2
```

```python
# 2개의 데이터 프레임을 좌우 열 방향으로 이어 붙이듯 연결하기 - r로 치면 cbind
result3 = pd.concat([df1,df2],axis=1) # axis = 1이면 cbind / 기본값은 0이다
result3
```

```python
result3_in = pd.concat([df1,df2],axis=1,join='inner') # join='inner' - 교집합
result3_in
```

```python
# Series 
sr1 = pd.Series(['e0','e1','e2','e3'],name='e')
sr2 = pd.Series(['f0','f1','f2'],name='f',index = [3,4,5])
sr3 = pd.Series(['g0','g1','g2','g3'],name='g')
print(sr1)
print(sr2)
print(sr3)
```

```python
# df1과 sr1을 좌우 열 방향으로 연결하기
result4 = pd.concat([df1,sr1],axis=1) 
result4
```

> df1과 sr1은 동일한 index를 사용함으로 무리없이 연결된다

```python
# df2 와 sr2 을 좌우 열방향으로 연결하기
result5 = pd.concat([df2,sr2],axis=1,sort=True)
result5
```

> index 2에 해당하는 data가 없어 NaN이 입력된다

```python
# sr1 과 sr3를 좌우 열방향으로 연결하기
result6 = pd.concat([sr1,sr3],axis=1)
result6
```

```python
# sr1 과 sr3를 상하 행방향으로 연결하기
result7 = pd.concat([sr1,sr3],axis=0)
result7
```

# DataFream 병합
- SQL의 join과 비슷한 방식으로 어떤 기준에 의해 병합하는 개념
- 이때 기준이 되는 열이나 index를 key라고 한다.
- key가 되는 열이나 index는 반드시 양쪽 DataFrame에 존재해야한다
- 사용 데이터 - stock price / valuation

```python
df1 = pd.read_excel('./Data/stock price.xlsx',engine='openpyxl')
df1.head()
```

> id : 종목코드   
> stock_name : 회사명   
> value : 시가 총액   
> price : 주가   

```python
df2 = pd.read_excel('./Data/stock valuation.xlsx',engine='openpyxl')
df2.head()
```

> id : 종목코드      
> name :회사명   
> eps : 주당 순이익   
> bps : 주당 순자산 가치   
> per : 주가 수익비율   
> pbr : 주가 자산비율   

```python
# 데이터 프레임 합치기 - 교집합
merge_inner = pd.merge(df1,df2)
merge_inner
```

```python
# 데이터 프레임 합치기 - 합집합
merge_outer = pd.merge(df1,df2,how='outer',on='id')
merge_outer
```

```python
# 데이터 프레임 합치기 - 왼쪽 데이터 프레임 기준
merge_left = pd.merge(df1,df2,how='left',left_on='stock_name',right_on='name')
merge_left
```

```python
# 데이터 프레임 합치기 - 오른쪽 데이터 프레임 기준
merge_right = pd.merge(df1,df2,how='right',left_on='stock_name',right_on='name')
merge_right
```

```python
# Booleam Indexing 으로 원하는 데이터 찾기
# df1 에서 price 가 50000미만 데이터 출력

mask = (df1['price'] < 50000)
price = df1.loc[mask,:]
price.head()
```

```python
value = pd.merge(price,df2)
value
```

```python
# value에서 name 컬럼 지우기

value.drop('name',axis=1,inplace=True)
value
```

# 데이터 프레임 결합
- join()은 merge()와 기본방식이 비슷하다.   
- join()은 두 데이터 프레임의 행 index를 기준으로 결합.   
- on = keys를 사용하면 행대신 열 가준으로 결합이 가능하다   

```python
df1.set_index(['id'],inplace=True)
df2.set_index(['id'],inplace=True)
```

```python
df1
```

```python
df2
```

```python
# 데이터프레임 결합
df3 = df1.join(df2)
df3.head()
```

```python
# 데이터 프레임 결합 : 교집합
df4 = df1.join(df2,how = 'inner')
df4
```
