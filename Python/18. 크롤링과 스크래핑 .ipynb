# 기본 사용법 

```python
from bs4 import BeautifulSoup

#HTML Sample
html = """
    <html>
        <body>
            <h1>Header</h1>
            <p>Line 1을 서술</p>
        </body>
    </html>
"""

#HTML 분석 
soup = BeautifulSoup(html,'html.parser') # HTML 내용 가져오기 
print(soup)

# 원하는 부분 출력
h1 = soup.html.body.h1
p1 = soup.html.body.p
print(h1)
print(p1)
#출력 
print("h1=",h1.string)
print("h1=",h1.text)
```

```python
# id를 요소로 추출하기 
from bs4 import BeautifulSoup
# HTML Sample 
html = """
    <html>
        <body>
            <h1 id = "title">Header</h1>
            <p id="body">Line 1을 서술</p>
        </body>
    </html>
"""

#HTML 분석 
soup = BeautifulSoup(html,'html.parser') # HTML 내용 가져오기 
print(soup)

# 원하는 부분 출력
h1 = soup.find(id = 'title')
p1 = soup.find(id = 'body')
print(h1)
print(p1)
#출력 
print("h1=",h1.string)
print("h1=",h1.text)
print("p1=",p1.string)
```

# 기상청 지료에서 필요한 데이터만 추출하기 

```python
import urllib.request
import urllib.parse
API = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
# 매개변수 URL 인코딩 
values ={'stdId':'109'}
params = urllib.parse.urlencode(values)
print(params)
# 요청 전용 URL 생성 
url = API + "?" + params
print(url)
data = urllib.request.urlopen(url).read()
#print(data)
# decoding하여 글자확인 
text = data.decode('utf-8')

#BeautifulSoup 분석 
soup = BeautifulSoup(text,"html.parser")
#print(soup)
# 원하는 데이터 추출 
title = soup.find('title').string 
wf = soup.find('wf').string

print(title)
#print(wf)
result =  ""
list_wfs = list(wf)
#print(list_wfs)
remove = ['<','b','r','/','>']

for lwf in list_wfs:
    if lwf in remove:
        continue
    else: 
        result += lwf
print(result.split('○')) # 쪼개서 다음줄로 넘긴다.
result = result.split("○")


for i in result:
    print(i)
```

# CSS 선택자 사용하기 
 - soup.select_one(): css선택자로 요소 하나를 추출 
 - soup.select(): css 선택자로 요소 여러개를 리스트로 추출

```python
from bs4 import BeautifulSoup

# 분석대상 HTML 
html = """
    <html>
        <body>
            <div id ="meigen">
                <h1>위키북스</h1>
                <ul class = "items">
                    <li>유니티 게임 이펙트 입문서</li>
                    <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
                    <li>모던 웹사이트 디자인의 정석</li>
                </ul>
            </div>
        </body>
    </html>
"""

# HTML 분석 
soup = BeautifulSoup(html, "html.parser")

# 필요한 부분 css로 출력 

# 타이틀 부분 출력하기 
h1 = soup.select_one("div#meigen > h1").string
#id: #, class:., tag: tag명, >:자손, ' ': 후손 

print(h1)
# 목록 부분 추출하기 
li_list = soup.select("div#meigen > ul.items > li")
print(li_list)
for li in li_list: 
    print(li.string)
```

# 네이버 금융에서 환율정보 추출하기 : https://finance.naver.com/marketindex/

```python
# 미국 환율 가져오기 
from bs4 import BeautifulSoup
import urllib.request as req 

# HTML 가져오기 
url = "https://finance.naver.com/marketindex/"
res = req.urlopen(url)

#HTML 분석하기 
soup = BeautifulSoup(res, "html.parser")

#데이터 추출하기 
price = soup.select_one("div.head_info > span.value")
print("usd/krw:",price)
#price = soup.select_one("#exchangeList > li.on > a.head.usd > div > span.value").string 
print(price)
```

```python
# 미국 / 일본 / 유럽연합 / 중국의 환율 
USA_price = soup.select_one("#exchangeList > li.on > a.head.usd > div > span.value").string 
JPN_price = soup.select_one("#worldExchangeList > li.on > a.head.jpy_usd > div > span.value").string
EU_price = soup.select_one("#exchangeList > li.on > a.head.eur > div > span.value")
China_price = soup.select_one("#exchangeList > li.on > a.head.cny > div.head_info.point_dn > span.value")
print("미국: ",USA_price)
print("일본: ",JPN_price)
print("유럽연합: ",EU_price)
print("중국: ", China_price)
```

# 다음 영화 예매율 
- https://movie.daum.net/ranking/boxoffice/weekly

```python
# 다음 많이 본 뉴스 추출 
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd


# HTML 가져오기 
url = "https://movie.daum.net/ranking/boxoffice/weekly"
res = req.urlopen(url)

#HTML 분석하기 
soup = BeautifulSoup(res, "html.parser")

movie_name = soup.select('.link_txt')[0:15]
count = 0 
titleLists = [] # 영화제목 리스트 
countLists = [] # 예매율 리스트

for li in movie_name:
    count += 1
    print(count ,":", li.string)
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# List 확인 
print(countLists)
print(titleLists)
```

```python
# countLists만으로 dataframe 생성 
df = pd.DataFrame(data=countLists, columns=['Index'])
df
```

```python
# 기존 데이터 프레임에 컬럼과 리스트 추가 
df['Titles'] = pd.Series(titleLists)
```

```python
# csv로 저장 
df.to_csv("./Data/DaumList.csv", sep=",", encoding='utf-8', index=False)
```
